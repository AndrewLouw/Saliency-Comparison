```python
import numpy as np
import seaborn as sns
import random
from keras.models import Model
from keras.layers import Dense, Input
from tqdm import tqdm
```

```python
value = 1
ground_truth = np.zeros([1000,1000])-value
for x in range(1000):
    for y in range(1000):
        x_scaled = x/1000
        y_scaled = y/1000
        if (x_scaled>=0.3) and (y_scaled>0.7):
            ground_truth[y][x]= value
        if (x_scaled<0.3) and (y_scaled>0.7):
            if (1-x_scaled < y_scaled):
                ground_truth[y][x]= value
        if (x_scaled>=0.6) and (y_scaled<0.4):
            ground_truth[y][x]= value
        if (x_scaled>0.3) and (y_scaled<0.2):
            if (0.5-x_scaled < y_scaled):
                ground_truth[y][x]= value
        if (x_scaled>0.3) and (0.2<=y_scaled<0.4):
            if (x_scaled*2/3 > y_scaled):
                ground_truth[y][x]= value
sns.heatmap(ground_truth)
```

```python
input_layer_1 = Input(shape = (2,))
classifier_1 = (Dense(2, activation='relu'))(input_layer_1)
model_1 = Model(input_layer_1,classifier_1)

input_layer_2 = Input(shape = (2,))
classifier_2 = (Dense(2, activation='relu'))(input_layer_2)
model_2 = Model(input_layer_2,classifier_2)

input_layer_3 = Input(shape = (2,))
classifier_3 = (Dense(2, activation='softmax'))(input_layer_3)
model_3 = Model(input_layer_3,classifier_3)

input_layer = Input(shape = (2,))
classifier = model_1(input_layer)
classifier = model_2(classifier)
classifier = model_3(classifier)
model = Model(input_layer,classifier)

model.compile(optimizer='adam', loss='binary_crossentropy')
model.save_weights("initial.h5")
```

```python
resolution = 100
input_dict = np.empty([resolution*resolution,2])
output_dict = np.empty([resolution*resolution,2])
for y in tqdm(range(resolution)):
    for x in range(resolution):
        sample = x+resolution*y
        input_dict[sample] = np.array([x/resolution,y/resolution])
        sample_y = int(y/resolution*1000)
        sample_x = int(x/resolution*1000)

        if np.array(ground_truth[sample_y][sample_x])==-1:
            output_dict[sample] = np.array([1,0])
        else:
            output_dict[sample] = np.array([0,1])
```

```python
best = 100
```

```python
samples = 4000
input_dict = np.empty([samples,2])
output_dict = np.empty([samples,2])

for test in tqdm(range(100)):
    for sample in range(samples):
        x = random.randrange(1000)
        y = random.randrange(1000)
        input_dict[sample] = np.array([x/1000,y/1000])
        if np.array(ground_truth[y][x])==-1:
            output_dict[sample] = np.array([1,0])
        else:
            output_dict[sample] = np.array([0,1])
    
    model.load_weights("initial.h5")
    train_history = model.fit(input_dict, output_dict,
                  epochs= 1600,
                  batch_size=samples,
                  validation_split = 0.20,
                  verbose = 0)
    if train_history.history['loss'][-1]<best:
        best = train_history.history['loss'][-1]
        model.save_weights("Xor_example.h5")
        print("new best: "+str(best))
    else: print("not good: "+ str(train_history.history['loss'][-1]) +" current: "+str(best))
```
